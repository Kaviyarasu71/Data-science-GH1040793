### **Introduction**

The Movies Dataset consists of 10,067 films providing detailed information across 15 features such as title, release year, runtime, genre, rating, metascore, synopsis, director, cast, votes, and box office gross. It brings together both audience perspectives through IMDb ratings and votes and critical evaluations that combine descriptive details like the cast and plot synopsis with metascore. This dataset can be used to analyze film trends, compare directors and genres, look at audience preferences, and even build recommendation systems.

### **Data Cleaning**

```python
df['votes'] = (df['votes'].astype(str).str.replace(',', '', regex=True) .replace('nan', '0').astype(int))
```

```python
cols = df.columns
cols = cols.str.strip()
cols = cols.str.lower()
cols = cols.str.replace(' ', '_')
cols = cols.str.replace('-', '_')
cols = cols.str.replace(r'[^a-z0-9_]', '', regex=True)
df.columns = cols
df['votes'] = df['votes'].astype(str)
df['votes'] = df['votes'].str.replace(',', '')
df['votes'] = df['votes'].str.extract(r'(\d+)')
df['votes'] = df['votes'].astype(float)
df['votes'] = df['votes'].astype('Int64')
```

```python
print(df['votes'].head())
```

```python

df['gross'] = df['gross'].astype(str).str.replace('$', '', regex=False)
df['gross'] = df['gross'].str.replace('M', 'e6', regex=False).str.replace('B', 'e9', regex=False)
df['gross'] = pd.to_numeric(df['gross'], errors='coerce')
```

```python
print(df['gross'].head())
```

```python
df_clean = df.dropna(subset=['metascore'])

df_clean.duplicated().sum()
```

```python
df_clean = df_clean.drop_duplicates()
```

```python
print("Before cleanup:\n", df.columns.tolist())
column_names = df.columns
column_names = column_names.str.strip()
column_names = column_names.str.lower()
column_names = column_names.str.replace(' ', '_')
column_names = column_names.str.replace('-', '_')
column_names = column_names.str.replace(r'[^a-z0-9_]', '', regex=True)
df.columns = column_names
print("\nAfter cleanup:\n", df.columns.tolist())


```

### **Data Visualization**

### **Business Questions**

#### 1.  What is the Gross revenue according to Genre ?

```python
df["primary_genre"] = df["genre"].str.split(",").str[0]

avg_gross_per_genre = df.groupby("primary_genre")["gross"].mean()

top_genres = avg_gross_per_genre.sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 5))
sns.barplot(x=top_genres.values, y=top_genres.index, palette="viridis")
plt.title("Average Gross Revenue by Genre (Top 10)")
plt.xlabel("Average Gross ($)")
plt.ylabel("Genre")
plt.tight_layout()
plt.show()


```

2. Who pulls the average gross revenue most?

```python


dir_stats = df.groupby("director")[["gross", "rating"]].mean()
dir_stats = dir_stats.dropna()
best_directors = dir_stats.sort_values(by="gross", ascending=False).head(15)


plt.figure(figsize=(8,4))
plt.barh(best_directors.index, best_directors["gross"])
plt.title("Directors Ranked by Avg Gross Revenue")
plt.xlabel("Mean Gross Earnings ($)")
plt.ylabel("Director")


plt.tight_layout()
plt.show()
```

3. Which director has most number of films?

```python
top_director = df['director'].value_counts().sort_values(ascending=False).head(10)
top_director
```

4. What is the average IMDB rating for every genre?

```python
average_ratings = df.groupby('genre')[['rating']].mean().sort_values(by='genre', ascending=False)
average_ratings.head(10)
genre_avg_ratings = df.groupby('genre')[['rating']].mean()
genre_avg_ratings = genre_avg_ratings.sort_values(by='rating', ascending=False)
genre_avg_ratings = genre_avg_ratings.head(10)
genre_avg_ratings.head(10)
```

5.  Which films are mostly considered as underrated?

```python
underrated = df[(df["metascore"]>80) & (df["gross"]<1e7)].sort_values("metascore", ascending=False)
underrated[["title","year","metascore","gross"]].head(15)
```

6. Which certificate is most helpful to earn both Revenue and Ratings together?

```python
df.groupby("certificate")[["gross","rating"]].mean().sort_values("gross", ascending=False)
```

7. Which movies had high critic scores but low revenue?

```python
high_score_low_gross = df[(df['metascore'] > 80) & (df['gross'] < 10_000_000)]
high_score_low_gross = high_score_low_gross.sort_values('metascore', ascending=False)
columns_to_see = ['title', 'year', 'metascore', 'gross']
print(high_score_low_gross[columns_to_see].head(10))


```

8. Which directors made most loved but financially underperforming films ?

```python
df[(df["metascore"]>80) & (df["gross"]<1e7)].groupby("director").size().sort_values(ascending=False).head(10)
```

9. Do movies with big stars earn more?

```python
cast_gross_df = df.melt(id_vars=["gross"],
value_vars=["cast1", "cast2", "cast3", "cast4"],
value_name="actor")

avg_gross_by_actor = cast_gross_df.groupby("actor")["gross"].mean()
top_10_actors = avg_gross_by_actor.sort_values(ascending=False).head(10)
top_10_actors.plot(kind="bar",
figsize=(6, 6),edgecolor="black" )
plt.title("Top 10 Actors by Average Gross")
plt.xlabel("Average Gross ($)")
plt.ylabel("Actor")
plt.tight_layout()
plt.show()
```
